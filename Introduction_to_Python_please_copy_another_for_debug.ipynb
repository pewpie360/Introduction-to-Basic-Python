# Variable Naming


> A variable can have a short name (like x and y) or a more descriptive name (age, carname, total_volume). Rules for Python variables:


*   A variable name must start with a letter or the underscore character
*   A variable name cannot start with a number
*   A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
*   Variable names are case-sensitive (age, Age and AGE are three different variables)





```python
#This is a good samples of naming a variable
This_Var = "Coin"
_MyVar_ = "Coin"
THISisVar = "Coin"
MyVar2 = "Coin"

#This is a bad samples of naming a variable
2VAR = "Flip"
this-var = "Flip"
My var = "Flip"
```

# Conditional Statements
### It is used in the anticipation of conditions occurring during the execution of a program and what specified actions should be taken according to the conditions


> There are 3 conditional types 


*   if statement
*   if ...elif... else statement
*   nested if statement





```python
x = 70
y = 35
if x > y:
  print("x is greater than y")

#The IF statement is similar to that of other languages. The if statement contains a logical expression using which the data is compared and a decision is made based on the result of the comparison.
if x < y:
  print("y is greater than x")
#The elif keyword is pythons way of saying "if the previous conditions were not true, then try this condition".
elif x > y:
  print("x is greater than y")

if x < y:
  print("y is greater than x")
elif x == y:
  print("x is equal to y")
#The else keyword catches anything which isn't caught by the preceding conditions.
else:
  print("x is greater than y")
```

```python
#nested if
#You can have if statements inside if statements, this is called nested if statements.
#There may be a situation when you want to check for another condition after a condition resolves to true.
#In such a situation, you can use the nested if construct. In a nested if construct, you can have an if...elif...else construct inside another if...elif...else construct. 
x = 50
if x > 10:
  print("x is above 10")
  if x > 20:
    print("x is also above 20")
  else:
    print("not above 20")
```
